# Data Visualization

```python
sns.boxplot(x="Classification",y="PL",data=data)
plt.ylabel("Petal Length")
```

```python
sns.boxplot(x="Classification",y="PW",data=data)
plt.ylabel("Petal Width")
```

```python
sns.stripplot(x="Classification",y="SL",data=data)
plt.ylabel("Sepal Length")
```

```python
sns.stripplot(x="Classification",y="SW",data=data)
plt.ylabel("Sepal Width")
```

```python
plt.figure(figsize=(17,10))
sns.boxplot(x="PW",y="PL",hue="Classification",data=data)
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
```

```python
plt.figure(figsize=(17,10))
sns.swarmplot(x="SW",y="SL",hue="Classification",data=data)
plt.xlabel("Sepal Width")
plt.ylabel("Sepal Length")
```
